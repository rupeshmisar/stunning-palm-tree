```python
# Import Required Library
import numpy as np
import pandas as pd
```

```python
# Load The file in Pandas Dataframe
df1 = pd.read_csv("amz_com-ecommerce_sample.csv" , encoding= 'unicode_escape')
df2 = pd.read_csv("flipkart_com-ecommerce_sample.csv",encoding= 'unicode_escape')
```

```python
# Check The shape of file
print(df1.shape)
print(df2.shape)
```

```python
df1.head(2)
```

```python
df2.head(2)
```

```python
# Merge both files On Colum uniq Id to Get Similar Product from both files

new_df = pd.merge(df1,df2,on='uniq_id',how = 'inner')
```

```python
new_df.head(2)
```

```python
new_df.shape
```

```python
final_df = new_df[['product_name_x','retail_price_x','discounted_price_x','product_name_y','retail_price_y','discounted_price_y']]
```

```python
# Rename the file as per Final Output
final_df.rename(columns =
              {'product_name_x': 'Product Name in flipkart','retail_price_x': 'retail_price in flipkart','discounted_price_x':'discounted_price in flipkart',
               'product_name_y': 'Product Name in Amazone','retail_price_y': 'retail_price in Amazone','discounted_price_y':'discounted_price in Amazone'}
              ,inplace=True)
```
